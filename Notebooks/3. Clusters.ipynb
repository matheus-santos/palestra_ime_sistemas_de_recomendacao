# 3. Clusters

```python
# Carregando tabelas users, reviews, places
# users: Alan, Barbara, Carlos, Denis, Edgar
# reviews: uid, pid, value
# places: 3 sushis, 3 pizzarias, 3 cantinas, 2 fast-food

%matplotlib inline

from sklearn.cluster import KMeans

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math
```

```python
# Carregando tableas
users = pd.read_csv("../data/users.csv")
reviews = pd.read_csv("../data/reviews.csv")
reviews_expanded = pd.read_csv("../data/reviews_expanded.csv")
places = pd.read_csv("../data/places.csv")

# Descrição de lugares
places.head()
```

```python
# Convertendo valores de string para numérico para manipulação matemática
# Cada categoria terá um id numérico e o mesmo será feito para o tipo de ambiente (`mood`)
places.loc[places.category == "Pizzaria", "category"] = 1
places.loc[places.category == "Comida Japonesa", "category"] = 2
places.loc[places.category == "Italiano", "category"] = 3
places.loc[places.category == "Fast-food", "category"] = 4

places.loc[places.mood == "Descontraido", "mood"] = 1
places.loc[places.mood == "Refinado", "mood"] = 2
places.loc[places.mood == "Negocios", "mood"] = 3

# Removendo features identificadoras `id` e `name`. 
# Não precisamos deste conteúdo por que não há como traçar correlação entre os itens do dataset
places = places.drop("id", 1)
places = places.drop("name", 1)

places.head()
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)
scatter = ax.scatter(places.avg_price, places.mood, c=places.category, s=50)

for i, j in np.random.randn(0, 50):
    ax.scatter(i, j, s=50, c="red", marker="+")

ax.set_xlabel("Avg_price")
ax.set_ylabel("Mood")

plt.colorbar(scatter)

fig.show()
```

```python
# Agrupando lugares baseado nas features `mood` e `avg_price`
est = KMeans(n_clusters=3)
est.fit(places[["mood", "avg_price"]])
labels = est.labels_

# Apresentando resultados
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)
scatter = ax.scatter(places.avg_price, places.mood, c=labels.astype(np.float), edgecolor='k')

for j, i in est.cluster_centers_:
    ax.scatter(i, j, s=50, c="red", marker="+")

ax.set_xlabel("Avg_price")
ax.set_ylabel("Mood")

plt.colorbar(scatter)

fig.show()
```

```python
# Gerando uma base maior
# Agrupando lugares baseado nas features `mood` e `avg_price`
np.random.seed(0)
X = np.random.random(size=(1000, 2)) * 10

# Agrupando lugares baseado nas features
est = KMeans(n_clusters=3)
est.fit(X)
labels = est.labels_

# Apresentando resultados
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)
scatter = ax.scatter(X[:, 0], X[:, 1], c=labels.astype(np.float), edgecolor='k')

for j, i in est.cluster_centers_:
    ax.scatter(i, j, s=50, c="red", marker="+")

ax.set_xlabel("Avg_price")
ax.set_ylabel("Mood")

plt.colorbar(scatter)

fig.show()
```
